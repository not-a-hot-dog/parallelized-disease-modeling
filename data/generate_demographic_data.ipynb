```python
import pandas as pd
import numpy as np
import pickle
```

```python
census_df = pd.read_csv('cc-est2018-alldata.csv', encoding='latin-1')
census_df
```

```python
census_df_latest = census_df[census_df.columns[:10]][census_df.YEAR==11]
census_df_latest
```

```python
combine_fips = lambda a, b: '{:0>2}{:0>3}'.format(a, b)
census_df_latest['FIPS'] = census_df_latest.STATE.combine(census_df_latest.COUNTY, combine_fips)
census_df_latest
```

```python
census_df_export = census_df_latest[['FIPS', 'AGEGRP', 'TOT_POP', 'TOT_MALE', 'TOT_FEMALE']]
census_df_export
```

```python
census_df_export.to_csv('census_data.csv', index=False, encoding='latin-1')
```
